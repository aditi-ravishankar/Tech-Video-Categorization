```python
import os
import re
import subprocess as s
import pandas as pd
```

### Read the data

```python
df = pd.read_csv('Data_Aditi - Sheet1.csv',header=[3])
df.head()
```

### Create the folder structure

```python
currentpath = os.getcwd()
print('The current path is : '+currentpath)
main_dir = ['unboxing','reviews','comparisons','others']
root_dir = 'Unboxing_Classifier_Data'
def direc():
    for i in range(0, len(main_dir)):
        dirname = currentpath+ '/' +root_dir + '/' + main_dir[i] 
        os.makedirs(dirname)
        print("Directory ", dirname, " Created ")
direc()
```

### Retrieve the transcripts from the YouTube videos

```python
for m in main_dir:

    links1=df[df['Type']==m]['Link']
    links1_list=links1.tolist()
    for i in range(0,len(links1_list)):
        URL=links1_list[i]
        outputfileloc = currentpath+ '/' + root_dir + '/' + m + "/" + "%(playlist)s/%(playlist_index)s_%(title)s.%(ext)s"
        cm = 'youtube-dl --write-sub --write-auto-sub --skip-download -o ' + outputfileloc + ' --no-check-certificate ' + URL
        p = s.Popen(cm.split(), stdout=s.PIPE, stderr=s.PIPE)
        o, e = p.communicate()
        p.wait()
        print("StdOut:\n", o)
        print("StdErr:\n", e)
```

### Clean the transcripts

```python
def preprocess_text(file_location):
    processed_text = []
    with open(file_location, "r") as f:
        for line in f.readlines()[3:]:
            if re.search(r"<c>|-->", line):
                pass
            else:
                processed_text.append(line.strip())
                
    return " ".join(processed_text)
```

```python
data_df=pd.DataFrame()
for label in os.listdir('Unboxing_Classifier_Data/'):
    for channel in os.listdir('Unboxing_Classifier_Data/'+label):
        for file_name in os.listdir('Unboxing_Classifier_Data/'+label+"/"+channel):
            file_location = os.getcwd() + "/" + "Unboxing_Classifier_Data/" + label + "/" + channel + "/"+file_name
            try:
                processed_text = preprocess_text(file_location)
            except:
                processed_text = ""
                print("error in processing file : ", file_location)
    
            data_df = data_df.append( 
            
                {
                    "label" : label,
                    "channel_name" : channel,
                    "file_name" : file_name,
                    "processed_text" : processed_text
                }, ignore_index = True
            
            )
                
```

### Remove punctuations

```python
import string
def pun(txt):
    punctuation_removed="".join([c for c in txt if c not in string.punctuation])
    return punctuation_removed
```

```python
data_df['processed_text']=data_df['processed_text'].apply(lambda x: pun(x))
```

### Tokenization

```python
from nltk.tokenize import word_tokenize
```

```python
data_df['prepared_text_tokenized']=data_df['processed_text'].apply(lambda x: word_tokenize(x))
```

### Remove Stopwords

```python
sw=nltk.corpus.stopwords.words('english')
def remove_stopwords(txt):
    removed=[word for word in txt if word not in sw]
    return removed
data_df['after_removing_stopwords']=data_df['prepared_text_tokenized'].apply(lambda x: remove_stopwords(x))
data_df
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import wordnet

def lemmatization(txt):
    def get_wordnet_pos(word):
        """Map POS tag to first character lemmatize() accepts"""
        tag = nltk.pos_tag([word])[0][1][0].upper()
        tag_dict = {"J": wordnet.ADJ,
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB,
                    "R": wordnet.ADV}

        return tag_dict.get(tag, wordnet.NOUN)
    lemmatizer = WordNetLemmatizer()
    lemmas=[]
    for word in txt:
        lemma = lemmatizer.lemmatize(word, get_wordnet_pos(word))                          
        lemmas.append(lemma)
                
    return lemmas

    
```

```python
data_df['after_lemmatization']=data_df['after_removing_stopwords'].apply(lambda x: lemmatization(x))
data_df
```

```python
data_df.to_csv('Final dataframe after lemmatization.csv') 
```

```python
df=data_df[['file_name','after_lemmatization','label']].copy()
df
```

```python
df['after_lemmatization']=df['after_lemmatization'].apply(lambda x: " ".join(x))
df
```

### TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
tr_f,tf,tr_o,to=train_test_split(df.after_lemmatization,df.label,test_size=0.3,random_state=42)
tfidf=TfidfVectorizer()
X_tfidf=tfidf.fit_transform(df.after_lemmatization)
X_train,X_test,y_train,y_test=train_test_split(X_tfidf,df.label,test_size=0.3,random_state=42)
X_tfidf.shape
```

### Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
NB = MultinomialNB().fit(X_train, y_train)
predicted=NB.predict(X_test)
df_final = pd.DataFrame()
df_final['lemmatized text']=tf
df_final['Actual label']=to
df_final['Prediction from Naive Bayes'] = predicted
df_final.head(10)
```

### Results of the Naive Bayes Classifier

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print('Accuracy : ',accuracy_score(y_test,predicted))
print(metrics.classification_report(y_test,predicted))
conf = confusion_matrix(y_test,predicted)
conf
```

### Support Vector Machine Classifier

```python
from sklearn import svm
sv = svm.SVC(kernel='linear')
sv.fit(X_train, y_train)
predicted2 = sv.predict(X_test)
```

```python
df_final['Prediction from SVM'] = predicted2
df_final.head(10)
```
